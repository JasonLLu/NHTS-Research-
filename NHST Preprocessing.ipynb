```python
num_features = ["HHSIZE"]
categ_features = ["DRIVER", "EDUC", "URBANSIZE", "URBRUR"]
all_features = num_features + categ_features + ["HHVEHCNT"]
# categ_map = {"DRIVER":["Skip","Yes","No"], 
#                     "EDUC":["idk","preferNotToAnswer","Skip","LessThanHighSchool","HighSchool","SomeCollegeOrAssociates","Bachelors","Graduates"], 
#                     "URBANSIZE" : ["50k-199k","200k-499k","500k-999k","1000kHeavyRail","1000kNoHeavyRail","NotUrban"],
#                      "URBRUR":["Urban","Rural"]}
# categ_map = {"DRIVER":[-1,1,0], 
#                     "EDUC":[-8,-7,-1,1,2,3,4], 
#                     "URBANSIZE" : [1,2,3,4,5],
#                      "URBRUR":[1,2]}
curr = data.copy(deep=True)[all_features]
curr.head()
```

```python
 for i, row in curr.iterrows():
    if curr.loc[i,"HHVEHCNT"] >= 3:
        curr.at[i,"HHVEHCNT"] = 3
```

```python
curr.head()
```

```python
for num_feature in num_features:
    normalizer = preprocessing.StandardScaler(with_mean = False)
    np_normalized = normalizer.fit_transform(curr.loc[:,num_feature].values.reshape(-1,1).astype(np.float32))
    
    curr.loc[:,num_feature] = np_normalized
curr.head()
```

```python
encoded_data = pd.get_dummies(data=curr, columns=categ_features)
```
