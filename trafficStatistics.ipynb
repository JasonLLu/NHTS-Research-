```python
#keep a copy for reuse later
reset = data.iloc[96:7667351,:].copy(deep=True)
```

```python
#let's use data2 for find the busiest times and zones
data2 = reset.iloc[96:7667351,:].copy(deep=True)
```

```python
data2['transactions'] = pd.Series(np.ones((len(data2))), index=data2.index)
```

```python
data2.tail()
```

```python
data2.set_index("tpep_pickup_datetime", inplace=True)
useful = ["transactions"]
```

```python
group_hour_sum = data2.groupby(pd.TimeGrouper(level='tpep_pickup_datetime', freq='1H'))[useful].agg("sum")
```

# group_hours_sum will tell you the number of transactions every hour

```python
group_hour_sum.head(24)
```

# Use this function to show only certain days of January binned by hour

```python
def showDayTransactions(day):
    #Day - Int: 1-31
    #Enter a day of month during January 
    i = (day - 1) * 24
    return group_hour_sum.iloc[i:i+24]
```

```python
showDayTransactions(1)
```

# Use this function to find the busiest hour of a day, and the number of transactions at that time

```python
def busiestHour(day):
    #Returns the busiest time of the day, and the number of transactions at that time
    i = (day - 1) * 24
    maxtime = group_hour_sum.iloc[i:i+24].idxmax()
    return day, maxtime, group_hour_sum.loc[maxtime,"transactions"]
```

```python
day, maxtime, num_transactions = busiestHour(1)
print("The busiest time on January", day, "is:", maxtime[0])
print("There were", num_transactions[0], "transactions")
```

# Use group_day_modes to look up the most popular PULocation and DOLocation and most popular payment_type

```python
useful_modes = ["PULocationID","DOLocationID","payment_type"]
group_day_modes = data2.groupby(pd.TimeGrouper(level='tpep_pickup_datetime', freq='1D'))[useful_modes].agg(pd.Series.mode)
```

# The rest of the code below just shows some statistics like mean, max, min
# The data is also binned by 5 min intervals here.

```python
data = reset.iloc[96:7667351,:].copy(deep=True)
```

```python
data.tail()
```

```python
data.set_index("tpep_pickup_datetime", inplace=True)
```

```python
useful = ["passenger_count","trip_distance","fare_amount", "tip_amount", "total_amount"]
```

```python
time_group_mean = data.groupby(pd.TimeGrouper(level='tpep_pickup_datetime', freq='5T'))[useful].agg("mean")
```

```python
len(time_group_mean)
```

```python
time_group_mean.head()
```

```python
#Passenger mean count every 5 min increments
plt.figure()
yvars = ["passenger_count"]
plt.plot(time_group_mean["passenger_count"])
plt.legend("passenger_count")
plt.ylabel('Number of passengers')
plt.xlabel('5 Min Increments')
plt.show()
```

```python
#Trip distance count every 5 min
plt.figure()
yvars = ["trip_distance"]
plt.plot(time_group_mean["trip_distance"])
plt.legend("trip_distance")
plt.ylabel('Distance')
plt.xlabel('5 Min Increments')
plt.show()
```

```python
#fare_amount	tip_amount	total_amount mean
plt.figure()
yvars = ["fare_amount","tip_amount","total_amount"]
for elt in yvars:
    plt.plot(time_group_mean[elt])
plt.legend(yvars)
plt.ylabel('Amount')
plt.xlabel('5 Min Increments')
plt.show()
```
